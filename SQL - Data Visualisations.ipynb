```python
%sql
SELECT 
    HOUR(tpep_pickup_datetime) AS pickup_hour,
    COUNT(*) AS total_trips,
    ROUND(SUM(fare_amount), 2) AS total_revenue,
    ROUND(AVG(fare_amount), 2) AS avg_fare
FROM big_data_workshop.gold_data.clean_df
GROUP BY pickup_hour
ORDER BY pickup_hour;
```

```python
%sql
SELECT 
    passenger_count,
    COUNT(*) AS trip_count,
    ROUND(AVG(fare_amount),2) AS avg_fare
FROM big_data_workshop.gold_data.clean_df
GROUP BY passenger_count
ORDER BY passenger_count;
```

```python
%sql
SELECT 
    CASE 
        WHEN trip_distance < 2 THEN '0-2 km'
        WHEN trip_distance < 5 THEN '2-5 km'
        WHEN trip_distance < 10 THEN '5-10 km'
        ELSE '10+ km'
    END AS distance_bucket,
    COUNT(*) AS total_trips,
    ROUND(AVG(fare_amount),2) AS avg_fare,
    ROUND(SUM(fare_amount),2) AS total_revenue
FROM big_data_workshop.gold_data.clean_df
GROUP BY distance_bucket
ORDER BY total_trips DESC;
```

```python
%sql
SELECT 
    DATE(tpep_pickup_datetime) AS trip_date,
    ROUND(SUM(fare_amount),2) AS daily_revenue,
    COUNT(*) AS total_trips
FROM big_data_workshop.gold_data.clean_df
GROUP BY trip_date
ORDER BY trip_date;
```

```python
%sql
SELECT 
    HOUR(tpep_pickup_datetime) AS pickup_hour,
    ROUND(SUM(fare_amount),2) AS total_revenue
FROM big_data_workshop.gold_data.clean_df
GROUP BY pickup_hour
ORDER BY total_revenue DESC
LIMIT 5;
```

```python
%sql
SELECT 
    ROUND(AVG(fare_amount / trip_distance),2) AS avg_fare_per_km
FROM big_data_workshop.gold_data.clean_df;
```

```python
%sql
SELECT *
FROM (
    SELECT 
        passenger_count,
        fare_amount,
        trip_distance,
        ROW_NUMBER() OVER (
            PARTITION BY passenger_count
            ORDER BY fare_amount DESC
        ) AS rank
    FROM big_data_workshop.gold_data.clean_df
) ranked
WHERE rank = 1;
```

```python
%sql
SELECT 
    passenger_count,
    ROUND(SUM(fare_amount),2) AS total_revenue,
    ROUND(
        SUM(fare_amount) * 100 /
        SUM(SUM(fare_amount)) OVER (),
    2) AS revenue_percentage
FROM big_data_workshop.gold_data.clean_df
GROUP BY passenger_count
ORDER BY revenue_percentage DESC;
```
